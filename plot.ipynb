```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei', 'Noto Sans CJK SC', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False   # Fix for minus sign rendering
```

```python
df=pd.read_csv("./NOAI2026原始notebook链接.csv")
df.head()
```

```python
problems=set(df["题目名称"].values)
problems
```

```python
ranks={'NOAI 2026 基于NVIDIA Isaac Sim的具身智能Sim2Real轨迹预测问题':136,
 'NOAI 2026 知乎场景下的用户意图识别':136,
 'NOAI 2026 积·和':136,
 'NOAI 2026 迷宫信息预测':136,
 'Total':136}

plot=plt.figure(4)
for problem in problems:    
    rows=(df[df["题目名称"]==problem])
    # display(rows)
    participants=len(rows)
    x=np.arange(participants)
    plt.title(problem)
    for col in ["A榜分数","B榜分数"]:
        scores = rows[col]
        y=scores
        print(f"mean {scores.mean():.2f} | median {scores.median():.2f} | stderr {scores.std():.2f}")
        rank=ranks[problem]
        print(rows.iloc[rank-1][col])
        plt.scatter([rank],[rows.iloc[rank-1][col]])
        plt.plot(x,y,label=col)
    # plt.plot(rows["A榜分数"],rows["B榜分数"],label="A-B")
    plt.legend()
    plt.show()
```
